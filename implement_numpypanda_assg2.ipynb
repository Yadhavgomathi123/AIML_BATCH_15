##data visualization using matpotib


```python
people=['Ann','Brandon','chen','David','Emily','Farook','Gagan','Hamish','Imran','Katherine','Lily']
age=[21,12,32,45,37,18,28,52,5,40,48,15]
weight=[55,35,77,68,70,60,72,69,18,65,82,48]
height=[160,135,170,165,173,168,175,159,105,171,155,158]

import matplotlib.pyplot as plt


plt.scatter(weight,height)
plt.title("Relationship between Weight and Height of patients")
plt.ylabel("Height in meters")
plt.xlabel("Weight of patients in Kgs")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))
plt.title("People's Weight in kgs", fontsize=16, fontstyle='italic')
people = ["aisu", "akshu", "aru", "abhi", "rockey"]
height = [170, 180, 160, 175, 185]
weight = [70, 80, 75, 85, 90]
# Main plot function 'bar'
plt.bar(people, height=weight, width=0.6, color='orange', edgecolor='k', alpha=0.6)
plt.xlabel("People", fontsize=15)
plt.xticks(fontsize=14, rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)", fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))

#Main plot function'hist'
plt.hist(weight, color='red', edgecolor='k', alpha=0.75, bins=5)

plt.title("Histogram of Patient Weights", fontsize=18)
plt.xlabel("Weight in kgs", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
